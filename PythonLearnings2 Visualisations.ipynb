```python
## Pie chart

#sizes values are cummulative

#explode : how much far the selected slice move far (like python)

#colors should be from only pie plot

#float format: autopct
```

```python
import matplotlib.pyplot as plt

labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame with name and age
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
        'Age': [15, 35, 45, 70, 60]}
df = pd.DataFrame(data)

# Set the size of the figure
plt.figure(figsize=(8, 6))

# Plot scatter plot with marker size based on age
df.plot.scatter(x='Name', y='Age', s=df['Age']*15, color='blue', marker='o', alpha=0.5)  # Increase marker size based on age

# Add labels and title
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Scatter Plot with Marker Size based on Age')

# Display the plot
plt.show()
```

## Seaborn

```python
import seaborn as sns

df= sns.load_dataset("tips")

df.head()
```

```python
df.dtypes
```

#tip is dependent feature and others are independent. You should be able to identify the dependent feature.

#all the values of tips are dependent on other features .Mostly Output is dependent feature

#corelation helps us find interrelationships. only interger /float 
value ranges from -1 to 1 always in co-relations, which we call coeficient



```python
df.corr()  #always gives float/int values. 
```

#if you see above table, it is positively corelated means as bill is increasing, tip is also increasing. same with size

```python
#to visualise above , we will use heat map:

sns.heatmap(df.corr())
```

if you have 2 indepenedent features , and corelation of those features are same or they are higher ,towards 3, towards 9 etc. then it is possible to use only one feature . This is useful in feature selection in EDA

# Joinplot

```python
sns.jointplot(x='tip',y='total_bill', data=df,kind='hex') # hex: hexagonal shape
```

This suggest that mainly tip is between 2$ 

and overall bill is around 10 to 20$. 



```python
sns.jointplot(x='tip',y='total_bill', data=df,kind='reg') #probability density function is getting displayed with the curvey line
```

# Categorical Plots

## CountPlot

```python
sns.countplot('sex',data=df)  #sex has 2 categories
```

```python
sns.countplot('day',data=df)  #people go out to eat on sat and sunday 
```

above plotted graphs were taking value on x axis default

If you give y axis specifically , the graph will be ploteed on y axis/horizontally

```python
sns.countplot(y='sex',data=df)
```

## BarPlot

Barplot and count plot are same but , to plot a barplot, we need to provide x and y values. But in countplot , either give x or y value

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

# Boxplot

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill', data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')  #it considers all the features .displyed in percentile 

#by default it will take only integer /float and not the categorical data
```

orient : "v" | "h", optional
    Orientation of the plot (vertical or horizontal). This is usually
    inferred from the dtype of the input variables, but can be used to
    specify when the "categorical" variable is a numeric or when plotting
    wide-form data.    

```python
sns.boxplot(data=df,x='total_bill',y='day',hue='smoker')
```

```python
sns.boxplot(data=df,x='total_bill',y='day')
```
